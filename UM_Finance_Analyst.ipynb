#### Calculate Market Share


```python
total_market_cap = data['Mar Cap - Crore'].sum()
data['Market Share %'] = (data['Mar Cap - Crore'] / total_market_cap) * 100
```

```python
data
```

#### Price-to-Sales (P/S) Ratio

```python
data['PS Ratio'] = data['Mar Cap - Crore'] / (data['Sales Qtr - Crore'] * 4) # Annualizing the quarterly sales
```

```python
data
```

#### Sales-to-Market Cap Efficiency

```python
data['Sales to Market Cap %'] = (data['Sales Qtr - Crore'] * 4 / data['Mar Cap - Crore']) * 100 # Annualizing the sales
```

```python
data
```

### Correlation between Market Cap and Quarterly Sales

```python
correlation = data[['Mar Cap - Crore', 'Sales Qtr - Crore']].corr()
```

```python
correlation
```

### Segment into Large-Cap, Mid-Cap, Small-Cap

```python
def categorize_market_cap(market_cap):
    if market_cap > 500000:
        return 'Large Cap'
    elif 100000 <= market_cap <= 500000:
        return 'Mid Cap'
    else:
        return 'Small Cap'

data['Market Cap Category'] = data['Mar Cap - Crore'].apply(categorize_market_cap)
segment_analysis = data.groupby('Market Cap Category').agg({'Mar Cap - Crore': 'sum', 'Sales Qtr - Crore': 'sum'})
```

```python
segment_analysis
```

```python
plt.figure(figsize=(10, 6))
plt.bar(data['Market Cap Category'], data['Mar Cap - Crore'], color='skyblue', label='Market Cap (Crore)')
plt.bar(data['Market Cap Category'], data['Sales Qtr - Crore'], color='salmon', alpha=0.7, label='Sales Qtr (Crore)')

plt.title('Market Cap and Sales by Market Cap Category')
plt.xlabel('Market Cap Category')
plt.ylabel('Value (Crore)')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.bar(segment_analysis.index, segment_analysis['Mar Cap - Crore'], color='skyblue', label='Market Cap (Crore)')
plt.bar(segment_analysis.index, segment_analysis['Sales Qtr - Crore'], color='salmon', alpha=0.7, label='Sales Qtr (Crore)')

plt.title('Aggregated Market Cap and Sales by Market Cap Category')
plt.xlabel('Market Cap Category')
plt.ylabel('Value (Crore)')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8, 8))
plt.pie(segment_analysis['Sales Qtr - Crore'], labels=segment_analysis.index, autopct='%1.1f%%', startangle=140, colors=['salmon', 'purple', 'yellow'])
plt.title('Sales Distribution by Category')
plt.show()
```

```python
aggregated_data = data.groupby('Market Cap Category').agg({
    'Mar Cap - Crore': 'sum',
    'Sales Qtr - Crore': 'sum'
}).reset_index()

# Bar Chart for Aggregated Market Cap and Sales
plt.figure(figsize=(12, 7))
plt.bar(aggregated_data['Market Cap Category'], aggregated_data['Mar Cap - Crore'], color='skyblue', label='Market Cap (Crore)')
plt.bar(aggregated_data['Market Cap Category'], aggregated_data['Sales Qtr - Crore'], color='salmon', alpha=0.7, label='Sales Qtr (Crore)')
plt.title('Aggregated Market Cap and Sales by Market Cap Category')
plt.xlabel('Market Cap Category')
plt.ylabel('Value (Crore)')
plt.legend()
plt.show()
```

```python

plt.figure(figsize=(12, 6))
sns.boxplot(x='Market Cap Category', y='Mar Cap - Crore', data=data)
plt.title('Distribution of Market Cap by Category')
plt.xlabel('Market Cap Category')
plt.ylabel('Market Cap (Crore)')
plt.show()

# Box Plot for Sales Distribution
plt.figure(figsize=(12, 6))
sns.boxplot(x='Market Cap Category', y='Sales Qtr - Crore', data=data)
plt.title('Distribution of Sales by Category')
plt.xlabel('Market Cap Category')
plt.ylabel('Sales Qtr (Crore)')
plt.show()
```

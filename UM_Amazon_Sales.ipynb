## Data Visualization

```python
units_sold_per_country = offline_sales.groupby('Country')['Units Sold'].sum()
units_sold_per_country.plot(kind='bar', figsize=(10, 6), color='skyblue')
plt.title('Total Units Sold per Country (Offline Sales)')
plt.xlabel('Country')
plt.ylabel('Units Sold')
plt.xticks(rotation=80)
plt.show()
```

#### Creating a bar chart that visualizes the total number of units sold per country for offline sales. 
- The bar chart will show which countries have the highest and lowest offline sales. Countries with taller bars[Mexico, Rwanda] have sold more units offline compared to those with shorter bars [Comoros,Mauritania].
- If you see certain countries with unexpectedly low sales, it might be worth investigating potential issues such as distribution problems, cultural mismatches, or economic factors that might be affecting sales.


```python
item_sales.plot(kind='bar', figsize=(10, 6), color='skyblue')
plt.title('Total Units Sold per Item (Offilne)')
plt.xlabel('Item Types')
plt.ylabel('Units Sold')
plt.xticks(rotation=80)
plt.show()
```

#### Creating a bar chart to visualize the total number of units sold for each item type in offline sales.
- The bar chart will display the total units sold for each item type. Items with taller bars are more popular, indicating higher sales, while those with shorter bars are less popular.
- Items with low sales might need attention. This could involve re-evaluating their pricing, marketing strategy, or even discontinuing them if they consistently perform poorly.
- Analyzing the data over time can help identify trends in consumer preferences. If certain items show increasing or decreasing sales trends, you can adjust your sales strategy accordingly.

```python
plt.figure(figsize=(8, 8))
data['Sales Channel'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90)
plt.title('Sales Channels Distribution')
plt.ylabel('')
plt.show()
```

- The pie chart will show how sales are distributed across different channels, such as online, offline. Each slice represents a different sales channel, with its size proportional to the number of sales made through that channel.
- The pie chart shows a relatively balanced distribution across multiple channels, it might suggest a well-diversified sales strategy that could be more resilient to changes in consumer behavior or market conditions.

```python
plt.figure(figsize=(14, 7))
sns.boxplot(data=data, x='Region', y='Unit Price')
plt.title('Unit Prices by Region')
plt.xticks(rotation=45)
plt.show()
```

- The boxplot shows the distribution of unit prices for each region. The box represents the interquartile range (IQR), which contains the middle 50% of the data, with the line inside the box representing the median price.
- The "whiskers" extend to the minimum and maximum values within 1.5 times the IQR, while points outside this range are plotted individually as potential outliers.
- This boxplot is an effective way to visualize and compare the distribution of unit prices across different regions. It provides valuable insights into regional pricing patterns, variability, and potential outliers. These insights can inform pricing strategies, identify anomalies, and guide decisions on market segmentation and regional marketing approaches.

```python
plt.figure(figsize=(14, 7))
sns.boxplot(data=online_sales, x='Region', y='Unit Price')
plt.title('Unit Prices by Region')
plt.xticks(rotation=45)
plt.show()
```

- This boxplot provides a detailed view of how unit prices vary across different regions in your online sales data. It highlights differences in pricing strategies, identifies potential outliers, and offers insights into regional price variability. These findings can be valuable for optimizing pricing strategies, understanding market dynamics, and tailoring sales approaches to different regions.

```python
# Creating the bar graph
channels = ['Offline', 'Online']
revenues = [total_revenue_offline, total_revenue_online]

plt.figure(figsize=(8, 6))
plt.bar(channels, revenues, color=['blue', 'green'])
plt.xlabel('Sales Channel')
plt.ylabel('Total Revenue')
plt.title('Total Revenue for Offline and Online Sales')
plt.show()
```

- This bar graph effectively illustrates the total revenue generated by offline and online sales channels, offering a straightforward visual comparison. The insights derived from this comparison can guide strategic decisions, such as where to focus marketing efforts, how to allocate resources, and which channels may need further development or support.
